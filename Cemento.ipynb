```r
datos <- read.csv("Concrete_Data_Yeh.csv", sep = ",")
str(datos)
```

```r
head(datos)
```

```r
summary(datos)
```

```r
# Calcular la proporción de valores faltantes para cada variable

prop_faltantes <- sapply(datos, function(x) mean(is.na(x)))

# Mostrar los resultados

names(prop_faltantes) <- names(datos)
print(prop_faltantes)
```

```r
par(mfrow = c(2, 2))

hist(datos$cement)
hist(datos$slag)
hist(datos$flyash)
hist(datos$water)

hist(datos$superplasticizer)
hist(datos$coarseaggregate)
hist(datos$fineaggregate)
hist(datos$age)

par(mfrow = c(1, 1))

hist(datos$csMPa)
```

```r
boxplot(datos$cement, datos$slag, datos$flyash, datos$water)
```

```r
density(datos$cement)
```

```r
par(mfrow = c(2, 2))

plot(datos$cement, datos$csMPa)
plot(datos$slag, datos$csMPa)
plot(datos$flyash, datos$csMPa)
plot(datos$water, datos$csMPa)

plot(datos$superplasticizer, datos$csMPa)
plot(datos$coarseaggregate, datos$csMPa)
plot(datos$fineaggregate, datos$csMPa)
plot(datos$age, datos$csMPa)

par(mfrow = c(1, 1))
```

```r
variables_independientes <- c("cement", "slag", "flyash", "water", "superplasticizer", "coarseaggregate", "fineaggregate", "age")

matriz_correlacion <- cor(datos[, variables_independientes])
print(matriz_correlacion)
```

```r
install.packages("gplots")
```

```r
library(gplots)

paleta <- heat.colors(length(variables_independientes))
heatmap(matriz_correlacion, col = paleta)

```
